```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv.imread("./images/green_img.jpg")
print(img.shape)
```

```python
%matplotlib inline
plt.imshow(img)
```

```python
# since images are read in BGR form in opencv we need to convert it to RGB form to display in matplotlib.image
img_rgb=cv.cvtColor(img,cv.COLOR_BGR2RGBA)
plt.imshow(img_rgb)
```

```python
# creating a min and a max threshold
min_th=np.array([0,220,0])
max_th=np.array([50,255,50])
```

```python
mask=cv.inRange(img,min_th,max_th)
plt.imshow(mask,cmap='gray')
```

```python
img[mask!=0]=[0,0,0]
plt.ims
```
